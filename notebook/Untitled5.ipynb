```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np

def scrapeBukaLapak(jumlahHalaman):
    product_list = []
    url = "https://www.bukalapak.com/c/perawatan-kecantikan/perawatan-wajah/masker-wajah-wanita"
    for page in range(0, jumlahHalaman):
        page = page + 1
        base_url = url+'?from=navbar_categories&source=navbar&page=' + str(page)
        # print(base_url)

        r = requests.get(base_url)
        soup = BeautifulSoup(r.text, "html.parser")

        produk_kecantikan = soup.find_all('div', class_="product-card")
       
        for item in produk_kecantikan:
            produk = { }
            product_name = item.find("a", {"class":"product__name"})
            produk['nama_produk'] = product_name.text.replace('\n', "").strip()  
            produk['url']= 'https://www.bukalapak.com' + str(product_name.get('href'))        
            produk['harga']= item.find("span", {"class":"amount"}).text.replace('\n', "").strip()
            # review
            product_review = item.find("a", {"class":"review__aggregate"})
            try:
                product_review = product_review.text
                produk['review'] = int(product_review)
            except:
                produk['review'] = 0


            product_list.append(produk)

    return pd.DataFrame.from_records(product_list)

dataset =scrapeBukaLapak(2)
dataset
```

```python
def scrapeLinkBerita(jumlahHalaman,tanggal):
    daftarLink = []
    for halaman in range(0, jumlahHalaman):
        halaman = halaman + 1
        base_url = 'https://news.detik.com/indeks/berita/' + str(halaman)+"?date="+tanggal
        #print(base_url)

        # Request URL and Beautiful Parser
        r = requests.get(base_url)
        soup = BeautifulSoup(r.text, "html.parser")

        berita_container = soup.find('ul', {'id': 'indeks-container'})
        berita = berita_container.find_all('li')
      
        for item in berita:
            x = item.find("a", href=True)
            daftarLink.append(x['href'])
      

    return daftarLink;

scrapeLinkBerita(3,'04/02/2019')
```
