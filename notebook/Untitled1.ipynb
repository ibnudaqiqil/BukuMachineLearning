```python
df.loc[df['Age'] > 70 ,["Age","PassengerId","Survived","Sex"]]
```

```python
#filtering data (yang umur diatas 60 tahun)
df[ df['Age'] > 60 ]
```

```python
#filtering data (Jenis Kelamin)
df[ df['Sex'] == "female" ]
```

```python
df[ (df['Age'] > 30) & (df['Sex'] == "female")]
```

```python
femaleOver30 = df[ (df['Age'] > 30) & (df['Sex'] == "female")]
femaleOver30.describe()
```
